```python
# drop the columns which are not needed 

df.drop(columns=['CountyId', 'IncomeErr', 'IncomePerCapErr',  
                'Hispanic', 'White', 'Black', 'Native', 'Asian', 'Pacific', 
                'Professional', 'Service', 'Office', 'Construction','Production',
                'Drive', 'Carpool', 'Transit', 'Walk', 'OtherTransp',
                'WorkAtHome', 'MeanCommute', 'PrivateWork', 'PublicWork', 'SelfEmployed', 'FamilyWork', ], inplace=True)
df.head()
```

### Review features and values

```python
df['State'].unique()
```

```python
df['State'].nunique()
```

* I am keeping District of Columbia as if it is a state to represent it, but removing Puerto Rico to focus on USA states

```python
df = df[df['State'] != 'Puerto Rico'].reset_index(drop=True)

df.head()
```

```python
df['County'].nunique()
```

* there are over 3000 counties in USA but here we have less

```python
df.rename(columns={'TotalPop': 'Population'}, inplace=True)
```

```python
# review population men and women values
df['check_population'] = df['Men'] + df['Women'] == df['Population']

df[~df['check_population']]
```

```python
df.drop(columns=[ 'check_population' ], inplace=True)
```

```python
df.rename(columns={
    'IncomePerCap': 'IncomePerCapita',
    'Men': 'MenPopulation',
    'Women': 'WomenPopulation',
    'Poverty': 'PovertyRate',
    'ChildPoverty': 'ChildPovertyRate',
    'Employed': 'EmployedPopulation',
    'Unemployment': 'UnemploymentRate'
}, inplace=True)
```

```python
df.head().T
```

```python
# check if VotingAgeCitizen is smaller than EmployedPopulation in any record, which shouldnt be

df[df['VotingAgeCitizen'] < df['EmployedPopulation']]
```

```python
# correct the EmployedPopulation based on VotingAgeCitizen and UnemploymentRate

df.loc[(df['State'] == 'Alaska') & (df['County'] == 'Aleutians East Borough'), 'EmployedPopulation'] = round(df['VotingAgeCitizen'] * (1 - (df['UnemploymentRate'] / 100)))
df.loc[(df['State'] == 'Alaska') & (df['County'] == 'Aleutians West Census Area'), 'EmployedPopulation'] = round(df['VotingAgeCitizen'] * (1 - (df['UnemploymentRate'] / 100)))
```

```python
df[df['County'] == 'Aleutians East Borough']
```

```python
df[df['County'] == 'Aleutians West Census Area']
```

```python
df.describe(include='all').T
```

```python
df[df['UnemploymentRate'] == 0]
```

```python
# correct EmployedPopulation where UnemploymentRate is 0

df.loc[df['UnemploymentRate'] == 0, 'UnemploymentRate'] = round(((df['VotingAgeCitizen'] - df['EmployedPopulation']) / df['VotingAgeCitizen']) * 100, 1)
```

```python
# check the extreme values for UnemploymentRate
df[df['UnemploymentRate'] > 50]
```

* 69,6 % UnemploymentRate for 2017 seems unrealistic and likely due to a data entry issue. 
* however, considering the data source specifies 919 employed individuals among 3,023 VotingAgeCitizens, it appears intentional for practice purposes, so I am keeping it.

```python
# calculate UnemployedPopulation
df['UnemployedPopulation'] = (df['VotingAgeCitizen'] * df['UnemploymentRate'] / 100).round().astype(int)

# calculate EmploymentRate
df['EmploymentRate'] = (df['EmployedPopulation'] / df['VotingAgeCitizen'] * 100).round(1)
```
